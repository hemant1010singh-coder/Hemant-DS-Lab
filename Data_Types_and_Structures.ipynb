Q3.What are the main differences between lists and tuples in Python
->these are the diffrence between list and tupes
A list is mutable, which means its elements can be changed, added, or removed after creation, whereas a tuple is immutable, meaning once it is created, its values cannot be modified. Lists are written using square brackets [], while tuples are written using parentheses (). In terms of performance, tuples are generally faster and more memory efficient than lists because they do not support modification, so Python does not need extra internal management for resizing or updating them. Lists use slightly more memory since they allow dynamic changes like appending and removing elements.

Q11.What advantages do dictionaries offer over lists for certain tasks ?
->Dictionaries offer several advantages over lists for certain types of tasks in Python. The main benefit is that dictionaries store data in the form of key and value pairs, which makes it easier to access information using a meaningful name instead of remembering index numbers. In a list, you have to know the exact position of an item to get it, but in a dictionary, you can directly use the key to access the value.

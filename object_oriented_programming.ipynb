```python

'''Q2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.'''

from abc import ABC,abstractmethod
import math
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
class Circle(Shape):
  def __init__(self,r):
    self.r=r
  def area(self):
    print(math.pi*self.r*self.r)
class Rectangle(Shape):
  def __init__(self,l,b):
    self.l=l
    self.b=b

  def area(self):
    print(self.l*self.b)


c1=Circle(2)
c1.area()
r1=Rectangle(5,9)
r1.area()
```

```python
'''Q3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.'''
class Vehical:
  def __init__(self,vehical_type):
    self.vehical_type=vehical_type
class Car(Vehical):
  def __init__(self,vehical_type,model):
    super().__init__(vehical_type)
    self.model=model
class ElectricCar(Car):
  def __init__(self,vehical_type,model,bettary_capacity):
    super().__init__(vehical_type,model)
    self.bettary_capacity=bettary_capacity
  def details(self):
    print(self.vehical_type)
    print(self.model)
    print(self.bettary_capacity)
e1=ElectricCar("car",412,"497Gh")
e1.details()



```

```python
'''5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance'''


class BankAccount:
  def __init__(self,balance):
    self.__balance=balance
  def deposit(self,amount):
    self.__balance=self.__balance+amount
    print(f"your deposited amount is  {amount}")
    print(f"Now your account balance is {self.__balance}")
  def withdraw(self,amount):
    if self.__balance>=amount:
      self.__balance=self.__balance-amount
      print(f"your withdrew  amount is  {amount}")
      print(f"Now your remaining balance is {self.__balance}")
    else:
      print("Indsufficient amount")
  def show_account_details(self):
    print(self.__balance)
acc=BankAccount(7000)
acc.deposit(3000)
acc.withdraw(2500)
acc.show_account_details()



```

```python
'''Q7.Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers.'''

class MathOperations:
  @classmethod
  def add_numbers(cls,x,y):
    return x+y
  @staticmethod
  def subtract_numbers(x,y):
    return x-y

class_result=MathOperations.add_numbers(5,9)
print(class_result)
static_result=MathOperations.subtract_numbers(49,32)
print(static_result)


```

```python
'''Q9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
fraction as "numerator/denominator".'''

class Fraction:
  def __init__(self,numerator,denominator):
    self.numerator=numerator
    self.denominator=denominator
  def __str__(self):
    return (str(self.numerator)+"/"+str(self.denominator))
f1=Fraction(5,6)
print(f1)

```

```python
'''Q10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.'''

class Vector:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other):
    x=self.x+other.x
    y=self.y+other.y

    return Vector(x,y)

  def __str__(self):
    return (str(self.x)+','+str(self.y))
v1=Vector(2,3)
v2=Vector(5,6)
v3=v1+v2
print(v3)


```

```python
'''Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.'''
class Employee:
  def __init__(self,h,r):
    self.h=h
    self.r=r
  def calculate_salary(self):
    return (self.h*self.r)
class Manager(Employee):
  def __init__(self,h,r,bonus):

    super().__init__(h,r)
    self.bonus=bonus
  def calculate_salary(self):
    return (self.h*self.r+self.bonus)
e=Employee(10,500)
print(e.calculate_salary())
m=Manager(10,500,6230)
print(m.calculate_salary())
```

```python
'''Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.'''

from abc import ABC,abstractmethod
class Animal(ABC):
  @abstractmethod

  def sound(self):
    pass
class Cow(Animal):
  def sound(self):
    print("cow say moo")
class Sheep(Animal):
  def sound(self):
    print("Sheep say baa")
c1=Cow()
c1.sound()
s1=Sheep()
s1.sound()

```

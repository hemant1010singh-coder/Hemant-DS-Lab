Q7.What are the advantages of using generators over regular functions?
Ans->The primary advantage of using generators over regular functions is their remarkable memory efficiency. While a regular function computes its entire result at once and stores it in memory—which can easily consume all your system's resources if the dataset is massive—a generator yields one value at a time only when it is explicitly requested. This concept, known as lazy evaluation, means the generator remembers its current state, paused execution point, and local variables between calls without needing to hold the entire sequence in memory. Because of this on-demand processing, generators are perfect for handling extremely large files, continuous data streams, or infinite sequences where a regular function would fail or slow down your program. Furthermore, writing a generator using the yield statement provides a much cleaner and simpler way to create custom iterators compared to building full Python classes from scratch.

```python
#Q4.Write a Python function that checks if a given number is prime or not from 1 #to 200.
def is_prime(n: int) -> bool:
    """Check if a number is prime."""
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Test the function for numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
    else:
        print(f"{num} is not prime")



```

```python
#Q5. Create an iterator class in Python that generates the Fibonacci sequence up #to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n          # total number of terms
        self.count = 0      # numbers generated so far
        self.a = 0          # first Fibonacci number
        self.b = 1          # second Fibonacci number

    def __iter__(self):
        return self         # return iterator object itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # stop iteration

        if self.count == 0:
            self.count += 1
            return self.a

        elif self.count == 1:
            self.count += 1
            return self.b

        else:
            c = self.a + self.b
            self.a = self.b
            self.b = c
            self.count += 1
            return c


# ------------------------
# Using the iterator
# ------------------------

n = 10
o1 = Fibonacci(n)

for x in o1:
    print(x)

```

```python
#Q7. Implement a generator function that reads a file line by line and yields each #line as a string.
def read_file_line_by_line(file_path):
    """
    Generator function जो file को line by line read करता है।
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                # yield keyword line को return करता है और function की state save रखता है
                yield line.strip()
    except FileNotFoundError:
        print("file not found")
```

```python

#Q8.Use a lambda function in Python to sort a list of tuples based on the second #element of each tuple
data = [(1, 30), (2, 10), (3, 50), (4, 20), (5, 40)]

# Ascending order (default)
sorted_asc = sorted(data, key=lambda x: x[1])
print(sorted_asc)
# [(2, 10), (4, 20), (1, 30), (5, 40), (3, 50)]

# Descending order
sorted_desc = sorted(data, key=lambda x: x[1], reverse=True)
print(sorted_desc)
# [(3, 50), (5, 40), (1, 30), (4, 20), (2, 10)]

# In-place sorting (modifies the original list)
data.sort(key=lambda x: x[1])
print(data)
# [(2, 10), (4, 20), (1, 30), (5, 40), (3, 50)]
```

```python
#Q9. Write a Python program that uses `map()` to convert a list of temperatures #from Celsius to Fahrenheit

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temps = [0, 10, 20, 30, 37, 40, 100]


fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from #a given string.
def is_not_vowel(char):
    return char.lower() not in "aeiou"

text = "Hello, World!"

filtered_chars = filter(is_not_vowel, text)
result = "".join(filtered_chars)

print(result)
```
